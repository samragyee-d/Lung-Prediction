```python
#Importing data and setting up X and y values 
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle 

#Defining columns to variables
lung_data = pd.read_csv('lung.csv')
lung_data.select_dtypes(exclude=['object'])
predict = "Level"

#Excluding "Level" (prediction) and "Patient Id" (irrelevant) 
X = np.array(lung_data.drop(columns = ["Level", "Patient Id", "index"]))
y = np.array(lung_data[predict])


#Establishing linear regression model
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.1)
linear = linear_model.LinearRegression()
linear.fit(X_train, y_train)
acc = linear.score(X_test, y_test)
print(acc)

#Determining "m" and "b" of linear equation or coeffiecnts of variables (24 for each feature)
print("Co: ", linear.coef_)
print("Intercept: ", linear.intercept_)

#linear Regression application 
predictions = linear.predict(X_test)
for x in range(len(predictions)):
    print(predictions[x], X_test[x], y_test[x])
```
